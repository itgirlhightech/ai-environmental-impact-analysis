```python
import pandas as pd 
import numpy as np

#leitura do arquivo csv / A.I csv
df = pd.read_csv('../data/raw/ai_adoption_dataset.csv')
df.head(3)
```

```python

#media adoção por ano
adocao = df.groupby("year")["adoption_rate"].mean().sort_index()
#diferença de um ano pro outro
tempo = adocao.diff()
print(tempo)
print(adocao)
```

```python
#analise por pais
paisAnalise = df.groupby("country")["adoption_rate"].mean().sort_values(ascending=False)
print(paisAnalise)
```

```python
#ranking de setores
setor = df.groupby("industry")["adoption_rate"].mean().sort_values(ascending=False)
print(setor.head(10))
```

```python
#pais por setor 
paisSetor = df.pivot_table(
    values = "adoption_rate",
    index = "industry",
    columns = "country",
    aggfunc = "mean"
)

print(paisSetor)
```

```python
import pandas as pd
import numpy as np

#leitura do arquivo csv de consumo de energia / csv eletricity consumption
df = pd.read_csv(r"../data/raw/MEC02.20260316203407.csv")
df.tail(3)
```

```python
#consumo ao longo do tempo
consumoTempo = df.groupby("TLIST(Q1)")["VALUE"].mean()
print(consumoTempo)
```

```python
# percentual do consumo de data centres ao loongo dos anos e dos valores
df_dc = df[df["Electricity Consumption"] == "Data centres"]
df_dc = df_dc.sort_values("TLIST(Q1)")
df_dc["growth"] = df_dc["VALUE"].pct_change()
print(df_dc[["TLIST(Q1)", "VALUE", "growth"]]) 
```

```python
#consumo por categoria
dfCategoria = df[df["Electricity Consumption"] == "Data centres"]
consumoCategoria = dfCategoria.groupby("TLIST(Q1)")["VALUE"].mean().reset_index()
print(consumoCategoria)
```

```python
consumoCategoria = dfCategoria.groupby("TLIST(Q1)")["VALUE"].mean().reset_index()
consumoCategoria.to_csv("../data/processed/dataset_cedt.csv", index=False)
```

```python
import pandas as pd

#leitura arquivo csv de cinsumo energico em escala global e preditiva
df = pd.read_csv(r"../data/raw/WEO2025_AnnexA_Free_Dataset_World.csv")
df.head(3)
```

```python
#consumo por ano
consumoAno = df.groupby("YEAR")["VALUE"].mean()
print(consumoAno)
```

```python
#comporação por cenario e ano
df = df.sort_values("YEAR")
comparaçãoC = df.groupby(["SCENARIO", "YEAR"])["VALUE"].mean().reset_index()
comparaçãoC["growthC"] = comparaçãoC.groupby("SCENARIO")["VALUE"].pct_change()
print(comparaçãoC[["SCENARIO", "YEAR", "growthC"]])
```

```python
#df final 
df_final = comparaçãoC[["YEAR", "VALUE", "SCENARIO", "growthC"]]
df_final.to_csv("../data/processed/dataset_anosfuturos.csv", index=False)
```
